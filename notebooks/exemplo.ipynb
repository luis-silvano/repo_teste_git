```python
from pyspark.sql import functions as F

path_in = '/mnt/rawzone/demo/producao'
df = spark.read.parquet(path_in)

df = df.filter(F.col('flag_ativo') == 1)
df = df.withColumn('KPI_EXEMPLO', F.col('valor') * 2)

out = '/mnt/consumezone/demo/producao/fato_exemplo'
df.write.mode('overwrite').format('parquet').save(out)
```
